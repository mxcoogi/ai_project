```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Model
from tensorflow.keras.layers import *
from tensorflow.keras.utils import to_categorical
import cv2

# CFAR-10 데이터 세트를 적재한다.
# 훈련 세트와 테스트 세트를 반환받는다. 
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
inputs = Input(shape=(32,32,3))
x = Conv2D(64, (3, 3), activation='relu')(inputs)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu')(x)
x = Flatten()(x)
x = Dense(80, activation='relu')(x)
outputs = Dense(10, activation='softmax')(x)
model = Model(inputs=inputs, outputs=outputs)

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
# 모델을 훈련한다. 
history = model.fit(x_train, y_train, epochs=10, verbose=1,
                  validation_split=0.3)

model.save('cifar10.keras')
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# RGB 이미지를 그레이스케일로 변환
x_train_gray = np.array([cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) for img in x_train])
x_test_gray = np.array([cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) for img in x_test])

# 그레이스케일 이미지를 3채널로 복제
x_train_gray_rgb = np.repeat(x_train_gray[..., np.newaxis], 3, -1)
x_test_gray_rgb = np.repeat(x_test_gray[..., np.newaxis], 3, -1)

# 데이터 정규화
x_train_gray_rgb = x_train_gray_rgb.astype('float32') / 255.0
x_test_gray_rgb = x_test_gray_rgb.astype('float32') / 255.0

# 타겟 데이터를 원-핫 인코딩
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
class_labels = [
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

# 테스트 이미지 중 하나 선택
for i in range(20):
    index = i
    test_image = x_test_gray_rgb[index]
    test_label = y_test[index]

    # 예측 수행
    prediction = model.predict(np.expand_dims(test_image, axis=0))
    predicted_class = np.argmax(prediction)
    print(test_label)

    # 결과 출력
    plt.imshow(test_image)
    plt.title(f'Predicted: {class_labels[predicted_class]}, Actual: {class_labels[np.argmax(test_label)]}')
    plt.show()
```
